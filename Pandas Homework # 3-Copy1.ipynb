```python
import pandas as pd
import matplotlib as plt
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
movies = pd.read_csv("imdb_1000.csv")
movies
```

```python
# calculate the average movie duration

avg_duration = movies.duration.mean()

round(avg_duration, 2)
    
 
```

```python
# sort the DataFrame by duration to find the shortest and longest movies

movies.sort_values('duration')
```

```python
# create a histogram of duration, choosing an "appropriate" number of bins

movies.hist(column='duration', bins=10)
```

```python
# use a box plot to display that same data

movies.boxplot(column='duration')
```

```python
# count how many movies have each of the content ratings

movies.content_rating.value_counts()
```

```python

# use a visualization to display that same data, including a title and x and y labels

a = movies.content_rating.value_counts()

x = list(a.index)
y = list(a)

fig, ax = plt.subplots()    
width = 0.75 # the width of the bars 
ind = np.arange(len(y))  # the x locations for the groups
ax.barh(ind, y, width, color="blue")
ax.set_yticks(ind+width/2)
ax.set_yticklabels(x, minor=False)
for i, v in enumerate(y):
    ax.text(v + .25, i + .25, str(v), color='blue', fontweight='bold') #add value labels into bar
plt.title('IMDB 1000 Ratings')
plt.xlabel('Number of Movies')
plt.ylabel('Movie Rating')
plt.show()
```

```python

# convert the following content ratings to "UNRATED": NOT RATED, APPROVED, PASSED, GP
movies.content_rating = movies.content_rating.str.replace('NOT_RATED','UNRATED')
movies.content_rating = movies.content_rating.str.replace('APPROVED','UNRATED')
movies.content_rating = movies.content_rating.str.replace('PASSED','UNRATED')
movies.content_rating = movies.content_rating.str.replace('GP','UNRATED')

```

```python

# convert the following content ratings to "NC-17": X, TV-MA

movies.content_rating = movies.content_rating.str.replace('X','NC-17')
movies.content_rating = movies.content_rating.str.replace('TV-MA','NC-17')
movies
```

```python

# count the number of missing values in each column

movies.isnull().sum()
```

```python

# count the number of missing values in each column


null_data = movies[movies.isnull().any(axis=1)]
null_data
```

```python
# if there are missing values inspect them and replace them with something reasonable
#where eagles dare is rated PG
# Butch Cassidy & Sundance Kid is rated PG
# True Grit is rated G


movies = movies.fillna({'content_rating': 'PG'}, inplace=True)
movies
# Test to see if it worked

print movies.loc[936]

# it worked
```

```python
x = movies['duration']
y = movies['star_rating']
plt.scatter(x, y)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

plt.show()
```
